```python
def term1(age):
    if age > 50:
        return 1.85*(age-50)
    else:
        return 0
```

```python
def term3(BMI, isAfrican, cigs):
    term = 1.17*BMI;
    if isAfrican :
        term = term + 22.34
    if cigs :
        term = term + 14.33;
    return term
```

```python
def term4(SBP, donor_sex, recipient_sex, ABO, related):
    term = 0.44*SBP;
    if donor_sex == 'male' and donor_sex == recipient_sex :
        term = term - 21.68;
    if ABO == 1:
        term = term + 27.30
    if related == 0:
        term = term -10.61
    return term
```

```python
def term7(donor_weight, recipient_weight):
    DRWR = donor_weight/recipient_weight;
    return min(DRWR, 0.9)
```
