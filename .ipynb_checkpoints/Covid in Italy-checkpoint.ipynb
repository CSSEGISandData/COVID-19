```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# sns.set_style('whitegrid')
plt.style.use('ggplot')
sns.set(rc={'figure.figsize':(10, 8)})
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
# pd.set_option('display.max_rows', 500)
```

```python
confirmed_df = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
death_df = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovery_df = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
columns_name = ['Confirmed', 'Death', 'Recovery']
```

```python
df_list = [confirmed_df, death_df, recovery_df]
```

```python
# Country list
# confirmed_df['Country/Region'].value_counts().sort_index()
```

```python
temp_list = []
i = 0
for df in df_list:
    df = df[df['Country/Region'] == country]
    df = df.groupby('Country/Region').sum()
    df = df.reset_index(drop=True)
    df = df.drop(df.columns.values[:2], axis=1)
    df = df.transpose()
    df = df.rename(columns={0:columns_name[i]})
    df.index = pd.to_datetime(df.index)
    temp_list.append(df)
    i = i + 1
df_list = temp_list
```

```python
full_df = pd.concat(df_list, axis=1)
```

```python
full_df.info()
```

```python
full_df['Confirmed Daily'] = full_df['Confirmed'] - full_df['Confirmed'].shift(1)
full_df['Death Daily'] = full_df['Death'] - full_df['Death'].shift(1)
full_df['Recovery Daily'] = full_df['Recovery'] - full_df['Recovery'].shift(1)
full_df.fillna(value=0, inplace=True)
full_df = full_df.astype('int64')
```

```python
full_df['Confirmed Daily pct'] = full_df['Confirmed Daily'] / full_df['Confirmed'].shift(1) * 100
full_df['Death Daily pct'] = full_df['Death Daily'] / full_df['Death'].shift(1) * 100
full_df['Recovery Daily pct'] = full_df['Recovery Daily'] / full_df['Recovery'].shift(1) * 100
full_df.fillna(value=0, inplace=True)
```

```python
# Фильтр по дате
full_df = full_df[full_df.index.month >= 2]
```

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.005)
fig.append_trace(go.Scatter(
    x=full_df.index,
    y=full_df['Confirmed'],
    mode='none',
    name='Total',
    fill='tozeroy'),
    row=1, col=1)

fig.append_trace(go.Bar(
    x=full_df.index,
    y=full_df['Confirmed Daily'],
    name='Daily'),
    row=2, col=1)

fig.update_layout(
    margin=dict(l=10, r=10, t=40, b=10), width=900, height=600, plot_bgcolor='rgb(245,245,245)',
    title= country + ' : Total vs Daily',
     font=dict(
#         family="Courier New, monospace",
        size=12,
        color="#7f7f7f"
    )
)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=full_df.index, y=full_df['Confirmed'], name='Confirmed', fill='tozeroy', mode='none')) # fill down to xaxis
fig.add_trace(go.Scatter(x=full_df.index, y=full_df['Death'], name='Death', fill='tozeroy', mode='none')) # fill to trace0 y
fig.add_trace(go.Scatter(x=full_df.index, y=full_df['Recovery'], name='Recovery', fill='tozeroy', mode='none')) # fill to trace0 y

fig.update_layout(
    margin=dict(l=10, r=10, t=40, b=10), width=1100, height=600, plot_bgcolor='rgb(245,245,245)',
    title=country + ' : Total Cases',
     font=dict(
#         family="Courier New, monospace",
        size=12,
        color="#7f7f7f"
    )
)
   
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=full_df.index, y=full_df['Confirmed Daily'], name='Confirmed')) # fill down to xaxis
fig.add_trace(go.Bar(x=full_df.index, y=full_df['Death Daily'], name='Death')) # fill to trace0 y
fig.add_trace(go.Bar(x=full_df.index, y=full_df['Recovery Daily'], name='Recovery')) # fill to trace0 y

fig.update_layout(
    margin=dict(l=10, r=10, t=40, b=10), width=1100, height=600, plot_bgcolor='rgb(245,245,245)',
    title=country + ' : Daily Cases',
     font=dict(
#         family="Courier New, monospace",
        size=12,
        color="#7f7f7f"
    ),
    barmode='stack'
)
fig.show()
```
